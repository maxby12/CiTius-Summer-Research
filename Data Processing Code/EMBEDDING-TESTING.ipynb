```python
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
x,y = np.random.uniform(low=10, high=11, size=100), np.random.uniform(low=10, high=11, size=100)
coords = list(zip(x,y))
coords = np.array(coords)
print(coords.shape)
print(coords)
```

```python
plt.scatter(coords[:,0], coords[:,1])
plt.show()
```

```python
mean, std = np.mean(coords, axis=0), np.std(coords, axis=0)

print(mean.shape, std.shape)
plt.scatter(coords[:,0], coords[:,1])
plt.scatter(mean[0], mean[1], c='g')
plt.scatter(std[0], std[1], c='r')
plt.show()
```

```python
candidates = np.apply_along_axis(lambda x: np.linalg.norm(x, ord=2),axis=1,arr=coords-mean)
print(candidates)
mean_distance, std_distance = np.mean(candidates), np.std(candidates)
print(mean_distance, std_distance)
print(np.linalg.norm(std, ord=2))
```
